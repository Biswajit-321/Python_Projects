#Analyze Economic Growth Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating DataFrames
economic_data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'GDP (in  billion)': [4500, 4815, 5210, 5616, 6071, 6480, 6364, 6963],
    'GDP Growth Rate (%)': [4.2, 7.0, 8.2, 7.8, 8.1, 6.7, -1.8, 9.5],
    'Per Capita Income (in )': [100000, 107000, 115000, 123000, 132000, 140000, 135000, 145000],
    'State Revenue (in  billion)': [500, 530, 580, 620, 670, 710, 700, 760],
    'Fiscal Deficit (% of GSDP)': [3.2, 3.0, 2.8, 2.5, 2.3, 2.2, 3.5, 2.1]
}

df_economic = pd.DataFrame(economic_data)

# Plotting GDP Growth Rate
plt.figure(figsize=(10, 5))
plt.plot(df_economic['Year'], df_economic['GDP Growth Rate (%)'], marker='o', linestyle='-')
plt.title('GDP Growth Rate of Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate (%)')
plt.grid(True)
plt.show()

# Plotting Per Capita Income
plt.figure(figsize=(10, 5))
plt.plot(df_economic['Year'], df_economic['Per Capita Income (in )'], marker='o', linestyle='-', color='g')
plt.title('Per Capita Income of Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (in )')
plt.grid(True)
plt.show()
```

#SUMMARY
Economic Growth:

GDP Growth: There was a steady increase in GDP from 4500 billion in 2014 to 6963 billion in 2021. Despite a decline in 2020 due to external factors, the GDP growth rate rebounded in 2021.

Per Capita Income: Increased consistently from 100,000 in 2014 to 145,000 in 2021.

State Revenue: Grew from 500 billion in 2014 to 760 billion in 2021.

Fiscal Deficit: Displayed a downward trend, indicating better fiscal management, except for a spike in 2020.

# Analyze Agricultural Data using Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Agricultural Data
agricultural_data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Major Crop Production (in tons)': [15000, 16000, 17000, 17500, 18000, 18500, 19000, 19500],
    'Irrigated Area (in hectares)': [100000, 110000, 120000, 130000, 140000, 150000, 160000, 170000],
    'Agricultural Income (in  billion)': [250, 270, 290, 310, 330, 350, 370, 390]
}

df_agricultural = pd.DataFrame(agricultural_data)

# Set the Seaborn style
sns.set(style="whitegrid")

# Plotting Major Crop Production
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Major Crop Production (in tons)', data=df_agricultural, marker='o', color='b')
plt.title('Major Crop Production in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Major Crop Production (in tons)')
plt.grid(True)
plt.show()

# Plotting Irrigated Area
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Irrigated Area (in hectares)', data=df_agricultural, marker='o', color='orange')
plt.title('Irrigated Area in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Irrigated Area (in hectares)')
plt.grid(True)
plt.show()

# Plotting Agricultural Income
plt.figure(figsize=(10, 5))
sns.lineplot(x='Year', y='Agricultural Income (in  billion)', data=df_agricultural, marker='o', color='g')
plt.title('Agricultural Income in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Agricultural Income (in  billion)')
plt.grid(True)
plt.show()

```

#SUMMARY
**Agricultural Development:**

Major Crop Production: Increased from 15,000 tons in 2014 to 19,500 tons in 2021.

Irrigated Area: Expanded from 100,000 hectares in 2014 to 170,000 hectares in 2021.

Agricultural Income: Grew from 250 billion in 2014 to 390 billion in 2021.

#SUMMARY
The analysis of industrial growth in Telangana indicates a thriving and dynamic industrial sector that has played a pivotal role in driving the state's economic development. #Analyze Social Indicators using Seaborn and Bar Plots

#SUMMARY
**Social Indicators:**

Population Growth: Decreased from 1.7% in 2014 to 1.2% in 2021, indicating a stabilizing population growth rate.

#Analyze Infrastructure Development and Education Data using Seaborn and Bar Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Infrastructure Development Data
infrastructure_data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Roads Built (km)': [1000, 1100, 1200, 1300, 1400, 1500, 1600, 1700],
    'Hospitals Built': [50, 55, 60, 65, 70, 75, 80, 85],
    'Educational Institutions Built': [100, 110, 120, 130, 140, 150, 160, 170]
}

df_infrastructure = pd.DataFrame(infrastructure_data)

# Education Data
education_data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'Primary Schools': [2000, 2100, 2200, 2300, 2400, 2500, 2600, 2700],
    'Secondary Schools': [1500, 1550, 1600, 1650, 1700, 1750, 1800, 1850],
    'Higher Education Institutions': [300, 310, 320, 330, 340, 350, 360, 370]
}

df_education = pd.DataFrame(education_data)

# Set the Seaborn style
sns.set(style="whitegrid")

# Plotting Roads Built
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Roads Built (km)', data=df_infrastructure, color='b')
plt.title('Roads Built in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Roads Built (km)')
plt.grid(True)
plt.show()

# Plotting Hospitals Built
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Hospitals Built', data=df_infrastructure, color='orange')
plt.title('Hospitals Built in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Hospitals Built')
plt.grid(True)
plt.show()

# Plotting Educational Institutions Built
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Educational Institutions Built', data=df_infrastructure, color='g')
plt.title('Educational Institutions Built in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Educational Institutions Built')
plt.grid(True)
plt.show()

# Plotting Primary Schools
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Primary Schools', data=df_education, color='b')
plt.title('Primary Schools in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Primary Schools')
plt.grid(True)
plt.show()

# Plotting Secondary Schools
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Secondary Schools', data=df_education, color='orange')
plt.title('Secondary Schools in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Secondary Schools')
plt.grid(True)
plt.show()

# Plotting Higher Education Institutions
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Higher Education Institutions', data=df_education, color='g')
plt.title('Higher Education Institutions in Telangana (2014-2021)')
plt.xlabel('Year')
plt.ylabel('Higher Education Institutions')
plt.grid(True)
plt.show()
```

#SUMMARY
**Education:**

Primary Schools:  Increased from 2,000 in 2014 to 2,700 in 2021.

Secondary Schools:  Grew from 1,500 in 2014 to 1,850 in 2021.

Higher Education Institutions:  Expanded from 300 in 2014 to 370 in 2021.

**Infrastructure Development:**

Roads Built: Extended from 1,000 km in 2014 to 1,700 km in 2021.

Hospitals Built: Increased from 50 in 2014 to 85 in 2021.

Educational Institutions Built: Grew from 100 in 2014 to 170 in 2021
